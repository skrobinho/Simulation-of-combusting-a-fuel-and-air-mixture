```python
from SDToolbox import *

# Symulation for various mixtures and temperatures

temperatures = [300, 450, 600, 750, 900]


#Calculation for CH4
f = open('CH4_post shock', 'w')
for i in temperatures:
    P = one_atm; Patm = P/one_atm
    T = i 
    air_fuel = 'CH4:1.0, O2:2.0, N2:7.52'
    mech = 'gri30_highT.cti'
    
    [cj_speed,R2] = CJspeed(P, T, air_fuel, mech, 0);   

    gas = PostShock_eq(cj_speed, P, T, air_fuel, mech)
    Ps = gas.P/one_atm
    
    initial_temperature = 'Calculations for %.d K\n' % i
    CJ_speed = 'CJ Speed is %.2f m/s\n' % cj_speed
    CJ_state = 'The CJ State is %.2f atm & %.2f K\n' % (Ps,gas.T)
    
    [equilibrium_sound_speed, frozen_sound_speed ] = equilSoundSpeeds(gas)
    
    sound_speeds = 'The sound speeds are: frozen sound speed = %.2f m/s & equilibrium sound speed = %.2f m/s\n\n' % (equilibrium_sound_speed, frozen_sound_speed)
    
    f.write(initial_temperature)
    f.write(CJ_speed)
    f.write(CJ_state)
    f.write(sound_speeds)

f.close()

#Calculation for C2H6
f = open('C2H6_post shock', 'w')
for i in temperatures:
    P = one_atm; Patm = P/one_atm
    T = i 
    air_fuel = 'C2H6:1.0, O2:3.5, N2:13.16'
    mech = 'gri30_highT.cti'
    
    [cj_speed,R2] = CJspeed(P, T, air_fuel, mech, 0);   

    gas = PostShock_eq(cj_speed, P, T, air_fuel, mech)
    Ps = gas.P/one_atm
    
    initial_temperature = 'Calculations for %.d K\n' % i
    CJ_speed = 'CJ Speed is %.2f m/s\n' % cj_speed
    CJ_state = 'The CJ State is %.2f atm & %.2f K\n' % (Ps,gas.T)
    
    [equilibrium_sound_speed, frozen_sound_speed ] = equilSoundSpeeds(gas)
    
    sound_speeds = 'The sound speeds are: frozen sound speed = %.2f m/s & equilibrium sound speed = %.2f m/s\n\n' % (equilibrium_sound_speed, frozen_sound_speed)
    
    f.write(initial_temperature)
    f.write(CJ_speed)
    f.write(CJ_state)
    f.write(sound_speeds)

f.close()   

#Calculation for C3H8
f = open('C3H8_post shock', 'w')
for i in temperatures:
    P = one_atm; Patm = P/one_atm
    T = i 
    air_fuel = 'C3H8:1.0, O2:5, N2:18.8'
    mech = 'gri30_highT.cti'
    
    [cj_speed,R2] = CJspeed(P, T, air_fuel, mech, 0);   

    gas = PostShock_eq(cj_speed, P, T, air_fuel, mech)
    Ps = gas.P/one_atm
    
    initial_temperature = 'Calculations for %.d K\n' % i
    CJ_speed = 'CJ Speed is %.2f m/s\n' % cj_speed
    CJ_state = 'The CJ State is %.2f atm & %.2f K\n' % (Ps,gas.T)
    
    [equilibrium_sound_speed, frozen_sound_speed ] = equilSoundSpeeds(gas)
    
    sound_speeds = 'The sound speeds are: frozen sound speed = %.2f m/s & equilibrium sound speed = %.2f m/s\n\n' % (equilibrium_sound_speed, frozen_sound_speed)
    
    f.write(initial_temperature)
    f.write(CJ_speed)
    f.write(CJ_state)
    f.write(sound_speeds)

f.close()  
```
